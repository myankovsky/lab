```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# ДАННЫЕ И ГРАФИК ЧИСТОГО ПАЛЬЦА

```python
clean_finger = pd.read_csv('Clean_Finger.csv', sep=';')
```

```python
clean_finger = clean_finger.apply(lambda x: x.str.replace(',','.'))
```

```python
wavelength = clean_finger['Lambda']
intensity = clean_finger['In']

wavelen = []
for i in range(len(wavelength)):
    wavelen.append(float(wavelength[i]))
    
inten = []
for i in range(len(intensity)):
    inten.append(float(intensity[i]))
```

```python
def annot_max(X1,Y1, ax=None):
    """Функция отрисовки максимсальных значений графика"""
    xmax = X1[np.argmax(Y1)]
    ymax = np.max(Y1)
    text= "Кальций λ={:.2f}, I={:.2f}".format(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94,0.96), **kw)
    
def annot_max_1(X1,Y1, ax=None):
    """Функция отрисовки максимсальных значений графика"""
    xmax = X1[np.argmax(Y1)]
    ymax = np.max(Y1)
    text= "Кальций λ={:.2f}, I={:.2f}".format(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text,xy=(xmax, ymax),  xytext=(0.935,0.92), **kw)  
```

```python
plt.figure(figsize=(16, 10), dpi= 200, facecolor='w')
#plt.subplot(figsize=())
plt.plot(wavelen, inten, linewidth=2)
plt.gca().set(xlim=(392, 400), ylim=(0.0, 11000))

plt.xlabel('Длина волны, нм', fontsize=20)
plt.ylabel('Интенсивность',fontsize=20)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Чистый палец', fontsize=20)  
    
annot_max(wavelen,inten)
annot_max_1(wavelen[3100:],inten[3100:])


plt.savefig('Clean_Finger.png')
```
